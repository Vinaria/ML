```python
import time
import numpy as np

import functions_vectorised
import functions
```

```python
#  

sum_non_neg_diag_data = []
for n in [10, 100, 1000, 10000]:
    X = np.random.randint(0, 3, (n, n))
    sum_non_neg_diag_data += [X]

are_multisets_equal_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(0, 2, n)
    y = np.random.randint(0, 2, n)
    are_multisets_equal_data += [(x, y)]

max_prod_mod_3_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 100, n)
    max_prod_mod_3_data += [x]

convert_image_data = []
for n in [10, 100, 1000]:
    image = np.random.random((n, n, 10))
    weights = np.random.random(10)
    convert_image_data += [(image, weights)]

rle_scalar_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 3, (n, 2))
    y = np.random.randint(1, 3, (n, 2))
    shapes = np.random.randint(1, 3, n)
    x[:, 1] = shapes
    y[:, 1] = shapes

    rle_scalar_data += [(x, y)]

cosine_distance_data = []
for n in [10, 100]:
    X = np.random.random((n, n))
    Y = np.random.random((n, n))
    cosine_distance_data += [(X, Y)]
```

```python
#     

def count_time(func, *data):
    tick = time.perf_counter()
    func(*data)
    return time.perf_counter() - tick
```

```python
#    

# sum_non_neg_diag

sum_non_neg_diag_time = []
sum_non_neg_diag_vectorised_time = []

for x in sum_non_neg_diag_data:
  sum_non_neg_diag_time += [count_time(functions.sum_non_neg_diag, x.tolist())]
  sum_non_neg_diag_vectorised_time += [count_time(functions_vectorised.sum_non_neg_diag, x)]

# are_multisets_equal
are_multisets_equal_time = []
are_multisets_equal_vectorised_time = []

for x in are_multisets_equal_data:
  are_multisets_equal_time += [count_time(functions.are_multisets_equal, x[0].tolist(), x[1].tolist())]
  are_multisets_equal_vectorised_time += [count_time(functions_vectorised.are_multisets_equal, x[0], x[1])]


# max_prod_mod_3
max_prod_mod_3_time = []
max_prod_mod_3_vectorised_time = []

for x in max_prod_mod_3_data:
  max_prod_mod_3_time += [count_time(functions.max_prod_mod_3, x.tolist())]
  max_prod_mod_3_vectorised_time += [count_time(functions_vectorised.max_prod_mod_3, x)]

# convert_image
convert_image_time = []
convert_image_vectorised_time = []

for x in convert_image_data:
  convert_image_time += [count_time(functions.convert_image, x[0].tolist(), x[1].tolist())]
  convert_image_vectorised_time += [count_time(functions_vectorised.convert_image, x[0], x[1])]


# rle_scalar
rle_scalar_time = []
rle_scalar_vectorised_time = []

for x in rle_scalar_data:
  rle_scalar_time += [count_time(functions.rle_scalar, x[0].tolist(), x[1].tolist())]
  rle_scalar_vectorised_time += [count_time(functions_vectorised.rle_scalar, x[0], x[1])]

# cosine_distance
cosine_distance_time = []
cosine_distance_vectorised_time = []

for x in cosine_distance_data:
  cosine_distance_time += [count_time(functions.cosine_distance, x[0].tolist(), x[1].tolist())]
  cosine_distance_vectorised_time += [count_time(functions_vectorised.cosine_distance, x[0], x[1])]
```

```python
data_size1 = np.array([10, 100, 1000, 10000])
data_size2 = np.array([10, 100, 1000])
data_size3 = np.array([10, 100])

f, axes = plt.subplots(nrows = 2, ncols = 3)
plt.subplots_adjust(right=1.5, top=1.3, wspace = 0.4, hspace = 0.4)
plt.setp(axes, xlabel = " ", ylabel = " ()")


ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[0, 2]
ax4 = axes[1, 0]
ax5 = axes[1, 1]
ax6 = axes[1, 2]

l1, = ax1.plot(data_size1, np.array(sum_non_neg_diag_time) * 1000, label=u"")
l2, = ax1.plot(data_size1, np.array(sum_non_neg_diag_vectorised_time) * 1000, label=u"")
ax1.set_title(u"sum_non_neg_diag")
#ax1.set_xticks(range(len(data_size1)))
#ax1.set_xlabels(range(len(data_size1)))
ax1.grid()
#ax1.legend()

ax2.plot(data_size1, np.array(are_multisets_equal_time) * 1000, label=u"")
ax2.plot(data_size1, np.array(are_multisets_equal_vectorised_time) * 1000, label=u"")
ax2.set_title(u"are_multisets_equal")
ax2.grid()
#ax2.legend()

ax3.plot(data_size1, np.array(max_prod_mod_3_time) * 1000, label=u"")
ax3.plot(data_size1, np.array(max_prod_mod_3_vectorised_time) * 1000, label=u"")
ax3.set_title(u"max_prod_mod_3")
ax3.grid()
#ax3.legend()

ax4.plot(data_size2, np.array(convert_image_time) * 1000, label=u"")
ax4.plot(data_size2, np.array(convert_image_vectorised_time) * 1000, label=u"")
ax4.set_title(u"convert_image")
ax4.grid()
#ax4.legend()

ax5.plot(data_size1, np.array(rle_scalar_time) * 1000, label=u"")
ax5.plot(data_size1, np.array(rle_scalar_vectorised_time) * 1000, label=u"")
ax5.set_title(u"rle_scalar")
ax5.grid()
#ax5.legend()

ax6.plot(data_size3, np.array(cosine_distance_time) * 1000, label=u"")
ax6.plot(data_size3, np.array(cosine_distance_vectorised_time) * 1000, label=u"")
ax6.set_title(u"cosine_distance")
ax6.grid()
#ax6.legend()

f.legend([l1, l2],["", ""])
f.show()
```
