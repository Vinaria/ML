````python
class KMeansClassifier(sklearn.base.BaseEstimator):
    def __init__(self, n_clusters):
        '''
        :param int n_clusters:            
        '''
        super().__init__()
        self.n_clusters = n_clusters
        self.kmeans_clusterizator = KMeans(self.n_clusters, n_init='auto')
        self.mapping = None

        #   :(   l|l)/

    def fit(self, data, labels):
        '''
               KMeans    ,    
        self._best_fit_classification   

        :param np.ndarray data:    -   
        :param np.ndarray labels:   .   .     -1.
                  .      
        :return KMeansClassifier
        '''
        self.kmeans_clusterizator.fit(data)
        cluster_labels = self.kmeans_clusterizator.predict(data)

        mapping, predicted_labels = self._best_fit_classification(cluster_labels, labels)

        self.mapping = mapping

        return self

    def predict(self, data):
        '''
              ,   .     
            1.      
            2.          

        :param np.ndarray data:    - 
        :return np.ndarray:   
        '''
        cluster_labels = self.kmeans_clusterizator.predict(data)
        predictions = self.mapping[cluster_labels]

        return predictions

    def _best_fit_classification(self, cluster_labels, true_labels):
        '''
        :param np.ndarray cluster_labels:   .   .
                [0, ..., n_clusters - 1]
        :param np.ndarray true_labels:   .   .
                -1.       .
                 
        :return
            np.ndarray mapping:         ,
                  mapping[idx] --     idx
            np.ndarray predicted_labels:     mapping  

                          
          .
            *       ,  ,      .
            *      ,          .
            *      ,       
        '''
        cluster_labels = np.array(cluster_labels)
        true_labels = np.array(true_labels)

        unique_classes, classes_counts = np.unique(true_labels[true_labels != -1], return_counts=True)
        biggest_class = unique_classes[classes_counts == classes_counts.max()][0]

        unique_clusters = np.unique(cluster_labels)
        mapping = np.empty(self.n_clusters)

        for i in range(self.n_clusters):
          classes_in_cluster = true_labels[cluster_labels == i]
          classes_in_cluster = classes_in_cluster[classes_in_cluster != -1]

          if not classes_in_cluster.size:
            mapping[i] = biggest_class
          else:

            unique, counts = np.unique(classes_in_cluster, return_counts=True)
            best_classes = unique[counts == counts.max()]
            mapping[i] = best_classes[0]


        predicted_labels = mapping[cluster_labels]

        return mapping, predicted_labels
````

````python
class MetaSemiSupervisedModel(sklearn.base.BaseEstimator):
    def __init__(self, data_processing, model, reducer_kwargs, model_kwargs):
        '''
        :param str data_processing:
             data_processing == 'UMAP',          
                  reducer_kwargs['n_components'].
            ,  reducer_kwargs['metric']      
                 
            ,    
        :param str model:
             model == 'LogisticRegression',       
                   model_kwargs['C']
             model == 'SelfTrainingClassifier',   SelfTrainingClassifier    LogisticRegression
                   model_kwargs['C']
             model == 'KMeansClassifier',   KMeansClassifier    model_kwargs['n_clusters']
        :param dict reducer_kwargs:
        :param dict model_kwargs:
        '''
        super().__init__()

        self.model = model
        self.data_processing = data_processing
        self.reducer_kwargs, self.model_kwargs = reducer_kwargs, model_kwargs

        self.scaler = StandardScaler()

        self.model = None
        self.reducer = None

        if data_processing == 'UMAP':
          if 'metric' in reducer_kwargs:
            self.reducer = UMAP(n_components=reducer_kwargs['n_components'], metric=reducer_kwargs['metric'])
          else:
            self.reducer = UMAP(n_components=reducer_kwargs['n_components'])

        match model:
          case 'LogisticRegression':
            self.model = LogisticRegression(C=model_kwargs['C'])
          case 'SelfTrainingClassifier':
            self.model = SelfTrainingClassifier(base_estimator=LogisticRegression(C=model_kwargs['C']))
          case 'KMeansClassifier':
            self.model = KMeansClassifier(model_kwargs['n_clusters'])

    def fit(self, data, labels):
        '''
        :param np.ndarray data:    -   .
             reducer_kwargs['metric'] == 'precomputed',  @data      
        :param np.ndarray labels:   .   .     -1.
                  .      
        :return MetaSemiSupervisedModel
        '''
        if self.reducer:
          data = self.scaler.fit_transform(data)
          data = self.reducer.fit_transform(data)
        self.model.fit(data, labels)

        return self

    def predict(self, data):
        '''
              ,   
        :param np.ndarray data:    - .
             reducer_kwargs['metric'] == 'precomputed',  @data    
                      .
        :return np.ndarray:   
        '''
        if self.reducer:
          data = self.scaler.transform(data)
          data = self.reducer.transform(data)
        predictions = self.model.predict(data)


        return predictions
````

````python
def plot_interactive(lowd_data, images, labels, names, n_dots=1000, image_scale=1.0):
    with matplotlib.rc_context(rc={
        'font.size': image_scale * matplotlib.rcParams['font.size'],
        'xtick.major.size': image_scale * matplotlib.rcParams['xtick.major.size'],
        'xtick.minor.size': image_scale * matplotlib.rcParams['xtick.minor.size'],
        'ytick.major.size': image_scale * matplotlib.rcParams['ytick.major.size'],
        'ytick.minor.size': image_scale * matplotlib.rcParams['ytick.minor.size'],

        'axes.linewidth': image_scale * matplotlib.rcParams['axes.linewidth'],
        'grid.linewidth': image_scale * matplotlib.rcParams['grid.linewidth'],
        'patch.linewidth': image_scale * matplotlib.rcParams['patch.linewidth'],
        'xtick.major.width': image_scale * matplotlib.rcParams['xtick.major.width'],
        'xtick.minor.width': image_scale * matplotlib.rcParams['xtick.minor.width'],
        'ytick.major.width': image_scale * matplotlib.rcParams['ytick.major.width'],
        'ytick.minor.width': image_scale * matplotlib.rcParams['ytick.minor.width'],

        'lines.markeredgewidth': image_scale * matplotlib.rcParams['lines.markeredgewidth'],
    }):
        fig, ax = plt.subplots(1, 1, figsize=(image_scale * 10, image_scale * 5))
        fig.set_dpi(300)
        ax.grid(True)

        n_clusters = len(np.unique(labels))

        scatter = plt.scatter(
            lowd_data[:n_dots, 0], lowd_data[:n_dots, 1], s=image_scale * 10,
            c=labels[:n_dots], cmap=plt.get_cmap('tab20', n_clusters), edgecolors='none'
        )

        cbar = plt.colorbar(scatter, ax=ax, label=' ')
        cbar.set_ticks(np.min(labels[:n_dots]) + (np.arange(n_clusters) + 0.5) * (n_clusters - 1) / n_clusters)
        cbar.set_ticklabels(names)

        offset_image = OffsetImage(images[0], zoom=image_scale * 2.0)
        ann_bbox = AnnotationBbox(
            offset_image, (0,0), xybox=(image_scale * 50., image_scale * 50.), xycoords='data',
            boxcoords="offset points", pad=0.3, arrowprops=dict(
                arrowstyle='->, head_length={0:.2f}, head_width={1:.2f}'.format(
                    image_scale * 0.4, image_scale * 0.2
                )
            )
        )
        ax.add_artist(ann_bbox)
        ax.set_title('  CIFAR10    2  ')
        ann_bbox.set_visible(False)

        def image_hover(event):
            if scatter.contains(event)[0]:
                ind, *_ = scatter.contains(event)[1]["ind"]
                w, h = fig.get_size_inches() * fig.dpi
                ws = (event.x > w / 2.) * -1 + (event.x <= w / 2.)
                hs = (event.y > h / 2.) * -1 + (event.y <= h / 2.)
                ann_bbox.xybox = (image_scale * 50.0 * ws, image_scale * 50.0 * hs)
                ann_bbox.set_visible(True)
                ann_bbox.xy =(lowd_data[ind, 0], lowd_data[ind, 1])
                offset_image.set_data(images[ind])
            else:
                ann_bbox.set_visible(False)
            fig.canvas.draw_idle()

        fig.canvas.mpl_connect('motion_notify_event', image_hover)

        plt.show()
````

````python
C_grid = np.linspace(1e-4, 10, 15)
n_clusters_grid = np.arange(1, 110, 9)
model_grid = ['LogisticRegression', 'SelfTrainingClassifier', 'KMeansClassifier']

total_results = []
best_params = []

for i, model_name in enumerate(model_grid):
  if i < 2:
    grid = C_grid
    param_name = 'C'
  else:
    grid = n_clusters_grid
    param_name = 'n_clusters'

  valid_results = []
  test_results = []

  for value in tqdm.tqdm(grid, desc=model_name):
    model = MetaSemiSupervisedModel(None, model_name, None, {param_name: value})
    model.fit(mnist_data_train, mnist_new_labels_train)
    valid_prediction = model.predict(mnist_data_valid)
    test_prediction = model.predict(mnist_data_new_test)
    valid_results.append(accuracy_score(mnist_labels_valid, valid_prediction))
    test_results.append(accuracy_score(mnist_labels_new_test, test_prediction))

  total_results.append([valid_results, test_results])
  best_params.append(grid[np.argmax(valid_results)])
````

````python
n_components_grid = np.arange(1, 110, 15)
model_grid = ['LogisticRegression', 'SelfTrainingClassifier', 'KMeansClassifier']

umap_total_results = []
umap_best_params = []

for i, model_name in enumerate(model_grid):
  if i < 2:
    param_name = 'C'
  else:
    param_name = 'n_clusters'
  param = best_params[i]

  valid_results = []
  test_results = []

  for value in tqdm.tqdm(n_components_grid, desc='UMAP+'+model_name):
    model = MetaSemiSupervisedModel('UMAP', model_name, {'n_components': value}, {param_name: param})
    model.fit(mnist_data_train, mnist_new_labels_train)
    valid_prediction = model.predict(mnist_data_valid)
    test_prediction = model.predict(mnist_data_new_test)
    valid_results.append(accuracy_score(mnist_labels_valid, valid_prediction))
    test_results.append(accuracy_score(mnist_labels_new_test, test_prediction))

  umap_total_results.append([valid_results, test_results])
  umap_best_params.append(n_components_grid[np.argmax(valid_results)])

````

````python
fig, (ax1, ax2) = plt.subplots(2, 3, figsize=(12, 8), layout='constrained')

for i, model in enumerate(model_grid):
  if i < 2:
    grid = C_grid
    param_name = 'C'
  else:
    grid = n_clusters_grid
    param_name = 'n_clusters'

  ax1[i].plot(grid, total_results[i][0], label='validation')
  ax1[i].plot(grid, total_results[i][1], label='test')
  line1 = Line2D([], [], marker='.', linestyle='None', label=f'best {param_name} = {round(best_params[i], 2)}')
  line2 = Line2D([], [], marker='.', linestyle='None', label=f'best test score = {round(max(total_results[i][1]), 3)}')
  ax1[i].add_line(line1)
  ax1[i].add_line(line2)
  ax1[i].set_title(model)
  ax1[i].set_xlabel(param_name)
  ax1[i].set_ylabel('accuracy')
  ax1[i].legend()

for i, model in enumerate(model_grid):
  grid = n_components_grid
  param_name = 'n_components'

  ax2[i].plot(grid, umap_total_results[i][0], label='validation')
  ax2[i].plot(grid, umap_total_results[i][1], label='test')
  line1 = Line2D([], [], marker='.', linestyle='None', label=f'best {param_name} = {round(umap_best_params[i], 2)}')
  line2 = Line2D([], [], marker='.', linestyle='None', label=f'best test score = {round(max(umap_total_results[i][1]), 3)}')
  ax2[i].add_line(line1)
  ax2[i].add_line(line2)
  ax2[i].set_title('UMAP + ' + model)
  ax2[i].set_xlabel(param_name)
  ax2[i].set_ylabel('accuracy')
  ax2[i].legend()
````
